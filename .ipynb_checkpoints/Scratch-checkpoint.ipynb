```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.model_selection import RandomizedSearchCV
import numpy as np


def svc_param_selection(X, y):
    Cs = [0.001, 0.01, 0.1, 1, 10]
    gammas = [0.001, 0.01, 0.1, 1]
    kernel = ['linear', 'poly', 'rbf']
    degree = [2,3,4,5]
    param_grid = {'C': Cs, 
                  'gamma' : gammas,
                  'kernel':kernel,
                  'degree':degree}
    rf_random = RandomizedSearchCV(SVC(),
                             scoring= 'average_precision',
                             param_distributions = param_grid,
                             cv=3,
                             n_iter=200,
                             verbose=False,
                             n_jobs=-1,
                             refit=False)
    
    rf_random.fit(X, y)
    return rf_random.best_params_

def rf_param_selection(X,y):
    n_estimators = [int(x) for x in np.linspace(start = 200, stop = 1000,num=1000-200+1)]
    param_grid = {'n_estimators': n_estimators}
    rf_random = RandomizedSearchCV(RandomForestClassifier(),
                               scoring = 'average_precision',
                               param_distributions = param_grid,
                               cv = 3,
                               n_iter=200,
                               verbose=False,
                               n_jobs = -1,
                               refit=False)

    rf_random.fit(X,y)
    return rf_random.best_params_

svc_params = svc_param_selection(clf.X,clf.label_bins[:,clf.label_position])
rf_params = rf_param_selection(clf.X,clf.label_bins[:,clf.label_position])
```
